```python
import MDAnalysis as mda
import os
from MDAnalysis.analysis import lineardensity as lin
import matplotlib.pyplot as plt
from math import sqrt
import math
%matplotlib inline
```

```python
current_directory = os.getcwd()
current_directory = str(current_directory)[:59]
data_files_directory_name = 'DataFiles'

graphene_water_psf = os.path.join(current_directory, data_files_directory_name, 'graphene_water.psf')
ns_graphene_water_dcd = os.path.join(current_directory, data_files_directory_name, '1ns.dcd')

#print(f'{graphene_water_psf} \n {ns_graphene_water_dcd}')
```

```python
u = mda.Universe(graphene_water_psf,ns_graphene_water_dcd)
```

```python
def NamesForAtoms(uni):
    names_of_atoms = []

    for atom in uni.atoms:

        junk1, unformatted_output = str(atom).split('resname ')
        formatted_output, junk2 = unformatted_output.split(',')

        if formatted_output not in names_of_atoms:
            names_of_atoms.append(formatted_output)
        else:
            pass

    return names_of_atoms
```

```python
uni_SPCE = u.select_atoms('resname SPCE')
pos = uni_SPCE.positions

x_pos = []
y_pos = []
z_pos = []

for coordinates in pos:
    x_pos.append(coordinates[0])
    y_pos.append(coordinates[1])
    z_pos.append(coordinates[2])

x_max = round(max(x_pos),2)
x_min = round(min(x_pos),2)
y_max = round(max(y_pos),2)
y_min = round(min(y_pos),2)
z_max = round(max(z_pos),2)
z_min = round(min(z_pos),2)


print(pos)
print('='*80)
print(x_pos)
print(y_pos)
print(z_pos)
print('='*80)
print(f'(Max, Min) | x_pos: ({round(float(x_max),2)}, {round(float(x_min),2)}) | y_pos: ({round(float(y_max),2)}, {round(float(y_min),2)}) | z_pos: ({round(float(z_max),2)}, {round(float(z_min),2)})')
# r = math.sqrt((x_pos)**2 + (y_pos)**2 + (z_pos)**2)
```

```python
polar_particle_coords = [] # x=a -> x=b as center line, positive direction is y is start of theta (x,r,theta) (A,A,Degree)
for coord in pos:

    x_coord = coord[0]
    radius = sqrt(coord[1]**2 + coord[2]**2)
    theta = math.acos(coord[1]/radius)*(360/(2*math.pi))
    polar_particle_coords.append([x_coord, radius, theta])

x_vals = []
r_vals = []
theta_vals = []

for i in range(len(polar_particle_coords)):
    x_vals.append(polar_particle_coords[i][0])
    r_vals.append(polar_particle_coords[i][1])
    theta_vals.append(polar_particle_coords[i][2])


greater_condition_radius = []
middle_condition_radius = []
lower_condition_radius = []
for element in r_vals:
    if element >= 70:
        greater_condition_radius.append(element)
    elif element >= 40:
        middle_condition_radius.append(element)
    elif element >= 10:
        lower_condition_radius.append(element)

v3 = math.pi * (max(greater_condition_radius)**2 - 70**2)
v2 = math.pi * (70**2 - 40**2)
v1 = math.pi * (40**2 - 10**2)

print(['Volume','Atoms'],[v3,len(greater_condition_radius)],[v2,len(middle_condition_radius)],[v1,len(lower_condition_radius)])
print(f'Density | 3:{round(len(greater_condition_radius)/v3,5)} | 2:{round(len(middle_condition_radius)/v2,5)} | 1:{round(len(lower_condition_radius)/v1,5)}')

#Density as a function of distance

a8075=[]
a7570=[]
a7065=[]
a6560=[]
a6055=[]

sus = []

for radius in r_vals:

    if 73>=radius and radius >= 72:
        sus.append(radius)

    if 80 <= radius and radius >= 75:
        a8075.append(radius)
    elif 75 <= radius and radius >= 70:
        a7570.append(radius)
    elif 70 <= radius and radius >= 65:
        a7065.append(radius)
    elif 65 <= radius and radius >= 60:
        a6560.append(radius)
    elif 60 <= radius and radius >= 55:
        a6055.append(radius)

print(len(sus) / (math.pi * (73**2 - 72**2))) 

#print(len(a8075) / (math.pi * (80**2-75**2)))
#print(len(a7570) / (math.pi * (75**2-70**2))) #Crazy
#print(len(a7065) / (math.pi * (70**2-65**2))) #Normal
#print(len(a6560) / (math.pi * (65**2-60**2)))
#print(len(a6055) / (math.pi * (60**2-55**2)))


#Volume = math.pi (max_radius**2 - min_radius**2)

''' 
When r >= 72, density begins to taper off pretty quickly, thus the radius of the best fit circle is 72A
x^2+y^2=r^2
x**2 + y**2 = r**2, r=72
'''
```
